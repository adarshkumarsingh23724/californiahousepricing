```python
#preparing the dataset
```

```python
dataset=pd.DataFrame(california_housing.data,columns=california_housing.feature_names)
```

```python
dataset.head()
```

```python
dataset['price']=california_housing.target
```

```python
dataset.head()
```

```python
dataset.info
```

```python
#summarizing the stats of data
dataset.describe()
```

```python
#check the missing value
dataset.isnull().sum()
```

```python
#exploratory data analysis
#correlation

dataset.corr()
```

```python
import seaborn as sns
sns.pairplot(dataset)
```

```python
plt.scatter(dataset['Population'],dataset['price'])
plt.xlabel('population')
plt.ylabel('price')
```

```python
import seaborn as sns
sns.regplot(x="Population",y="price",data=dataset)
```

```python
#independent and dependent feature

X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
#Train Test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
#standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
X_train
```

```python
X_test
```

```python
#model training
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
#print the coeffiecients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
##on which parameter the model has been trained
regression.get_params()
```

```python
### prediction with test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
###plot a scatter plot for the prediction
plt.scatter(y_test,reg_pred)
```

```python
##residuals
residuals=y_test-reg_pred
```

```python
residuals
```

```python
##plot this residuals

sns.displot(residuals,kind="kde")
```

```python
##scatter plot with respect to prediction and residuals
#uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test,reg_pred))
print(mean_squared_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
#r square and r adjusted square
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
#display adjusted r square
```

```python
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
